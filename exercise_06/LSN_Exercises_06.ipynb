In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

### Solution

In this simulation, a one-dimensional Ising model is explored using the Monte Carlo method across a range of temperatures from T = 2 to T = 0.5. The system is initially prepared with a random configuration of spins (half up and half down) as if starting at infinite temperature. The simulation begins at T = 2, allowing the system to equilibrate. Once equilibrium is reached, the temperature is gradually lowered in steps of 0.1 (e.g., from T = 2 to T = 1.9, and so on) without additional equilibration at each step, since the system remains close to equilibrium at nearby temperatures. This process is repeated until the final temperature of T = 0.5 is reached. 

A Monte Carlo step consists in trying to flip one time all the spins of the system.

#### Equilibration

The figure below shows the equilibration process at T = 2, where the system stabilizes before starting the temperature reduction. With this plot, the goal is to evaluate how many Monte Carlo steps are needed for equilibration.

```python
energy_eq = np.loadtxt('M(RT)^2/OUTPUT/EQUILIBRATION/total_energy.dat')
C_eq = np.loadtxt('M(RT)^2/OUTPUT/EQUILIBRATION/specific_heat.dat')
susceptibility_eq = np.loadtxt('M(RT)^2/OUTPUT/EQUILIBRATION/susceptibility.dat')
magnetization_eq = np.loadtxt('M(RT)^2/OUTPUT/EQUILIBRATION/magnetization.dat')

energy_eq_Gibbs = np.loadtxt('GIBBS/OUTPUT/EQUILIBRATION/total_energy.dat')
C_eq_Gibbs = np.loadtxt('GIBBS/OUTPUT/EQUILIBRATION/specific_heat.dat')
susceptibility_eq_Gibbs = np.loadtxt('GIBBS/OUTPUT/EQUILIBRATION/susceptibility.dat')
magnetization_eq_Gibbs = np.loadtxt('GIBBS/OUTPUT/EQUILIBRATION/magnetization.dat')

T = 2.
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
h = 0.02
l1 = np.exp(beta*J)*np.cosh(beta*h)+np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
l2 = np.exp(beta*J)*np.cosh(beta*h)-np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))
Z = l1**Ns + l2**Ns
M = (np.exp(beta*J)*np.sinh(beta*h)*((l1**(Ns-1))*(1+np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J))) 
        + (l2**(Ns-1))*(1-np.exp(beta*J)*np.cosh(beta*h)/np.sqrt(np.exp(2*beta*J)*np.cosh(beta*h)*np.cosh(beta*h)-2*np.sinh(2*beta*J)))))/(Z)
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)


limit = 2000

fig, ax = plt.subplots(2, 2, figsize=(16, 11))
fig.suptitle('Metropolis and Gibbs Equilibration', fontsize=30)
# ax[0, 0].plot(energy_eq[:limit,0], energy_eq[:limit,1], 'o', alpha=0.5)
ax[0,0].plot(np.cumsum(energy_eq[:limit,1]) / np.arange(1, len(energy_eq[:limit,1])+1), label='MC', lw =4)
ax[0,0].plot(np.cumsum(energy_eq_Gibbs[:limit,1]) / np.arange(1, len(energy_eq_Gibbs[:limit,1])+1), label='GIBBS', lw =4)
# ax[0,0].plot(energy_eq[window_size-1:,0], moving_average(energy_eq[:,1], window_size), label='Moving Average')
ax[0,0].axhline(y=e, color='k', linestyle='--', label='Expected Value')
ax[0,0].set_xlabel('MC Steps')
ax[0,0].set_ylabel('Energy')
ax[0,0].legend()

# ax[0, 1].plot(C_eq[:limit,0], C_eq[:limit,1], alpha=0.5)
ax[0,1].plot(np.cumsum(C_eq[:limit,1]) / np.arange(1, len(C_eq[:limit,1])+1), label='MC', lw =4)
ax[0,1].plot(np.cumsum(C_eq_Gibbs[:limit,1]) / np.arange(1, len(C_eq_Gibbs[:limit,1])+1), label='GIBBS', lw =4)
# ax[0,1].plot(C_eq[window_size-1:,0], moving_average(C_eq[:,1], window_size), label='Moving Average', lw =3)
ax[0,1].axhline(y=heat, color='k', linestyle='--', label='Expected Value')
ax[0,1].set_xlabel('MC Steps')
ax[0,1].set_ylabel('Specific Heat')
ax[0,1].legend()

# ax[1, 0].plot(susceptibility_eq[:limit,0], susceptibility_eq[:limit,1], alpha=0.5)
ax[1,0].plot(np.cumsum(susceptibility_eq[:limit,1]) / np.arange(1, len(susceptibility_eq[:limit,1])+1), label='MC', lw =4)
ax[1,0].plot(np.cumsum(susceptibility_eq_Gibbs[:limit,1]) / np.arange(1, len(susceptibility_eq_Gibbs[:limit,1])+1), label='GIBBS', lw =4)
# ax[1,0].plot(susceptibility_eq[window_size-1:,0], moving_average(susceptibility_eq[:,1], window_size), label='Moving Average', lw =3)
ax[1,0].axhline(y=X, color='k', linestyle='--', label='Expected Value')
ax[1,0].set_xlabel('MC Steps')
ax[1,0].set_ylabel('Susceptibility')
ax[1,0].legend()

# ax[1, 1].plot(magnetization_eq[:limit,0], magnetization_eq[:limit,1], alpha=0.5)
ax[1,1].plot(np.cumsum(magnetization_eq[:limit,1]) / np.arange(1, len(magnetization_eq[:limit,1])+1), label='MC', lw =4)
ax[1,1].plot(np.cumsum(magnetization_eq_Gibbs[:limit,1]) / np.arange(1, len(magnetization_eq_Gibbs[:limit,1])+1), label='GIBBS', lw =4)
# ax[1,1].plot(magnetization_eq[window_size-1:,0], moving_average(magnetization_eq[:,1], window_size), label='Moving Average', lw =3)
ax[1,1].axhline(y=M, color='k', linestyle='--', label='Expected Value')
ax[1,1].set_xlabel('MC Steps')
ax[1,1].set_ylabel('Magnetization')
ax[1,1].legend()

plt.tight_layout()
plt.show()
```

Based on the simulation cumulative average, which tends to reduce the raw thermodynamics quantities fluctuations and provides a more reliable indication of whether the sampling has stabilized, the system for $M(RT)^2$ and GIBBS Monte Carlo method appears to quickly converge after **1000 step** to the correct values at T = 2. 


#### Simulation

The following plots show the results of the Monte Carlo simulations compared with the corresponding theoretical predictions. For each temperature, the simulation was run independently, and the final value of the observable and its uncertainty were taken from the last block of the simulation. For each T the data were obtained by dividing the simulation into 40 blocks, with a total of $10^5$ Monte Carlo steps. 

1. $U(N,T) = \langle H \rangle_{h=0}$

```python
points=100
T = np.linspace(0.2,3.0,num=points)

def e(T):
    beta = 1/T
    J = 1.0
    Ns = 50
    th = np.tanh(J/T)
    thN= th**Ns
    ch = 1/th
    e = -J*( th + ch*thN )/( 1 + thN )
    return e

en = e(T)
energy_MC = np.loadtxt('M(RT)^2/OUTPUT/total_energy.dat')
energy_GIBBS = np.loadtxt('GIBBS/OUTPUT/total_energy.dat')



plt.figure(figsize=(10, 6))
# fig = plt.figure(figsize=(15, 5))

# plt.subplot(1, 2, 1)
plt.errorbar(energy_MC[:,0], energy_MC[:,1], yerr=energy_MC[:,2], fmt = "ko", ms = 10, label='MC simulation')
plt.errorbar(energy_GIBBS[:,0], energy_GIBBS[:,1], yerr=energy_GIBBS[:,2],fmt = "rx", ms = 10,mew=2, label='Gibbs simulation')
plt.plot(T, en, lw=5, label = 'Analytical solution')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()



plt.show()
```

2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

C_MC = np.loadtxt('M(RT)^2/OUTPUT/specific_heat.dat')
C_GIBBS = np.loadtxt('GIBBS/OUTPUT/specific_heat.dat')
plt.figure(figsize=(10, 6))
plt.errorbar(C_MC[:,0], C_MC[:,1], yerr=C_MC[:,2], fmt = "ko", ms = 10, label='MC simulation')
plt.errorbar(C_GIBBS[:,0], C_GIBBS[:,1], yerr=C_GIBBS[:,2], fmt='rx', ms=10, mew=2, label='Gibbs simulation')
plt.plot(T, heat, lw=5, label='Analytical solution')
plt.legend()
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.show()
```

3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$

```python
h=0.02 #external field
points=100
T = np.linspace(0.2,3.0,num=points)
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

magnetization_MC = np.loadtxt('M(RT)^2/OUTPUT/magnetization.dat')
magnetization_GIBBS = np.loadtxt('GIBBS/OUTPUT/magnetization.dat')
plt.figure(figsize=(10, 6))
plt.errorbar(magnetization_MC[:,0], magnetization_MC[:,1] , magnetization_MC[:,2], fmt = "ko", ms = 10, label='MC simulation')
plt.errorbar(magnetization_GIBBS[:,0], magnetization_GIBBS[:,1] , magnetization_GIBBS[:,2], fmt='rx', ms=10, mew=2, label='Gibbs simulation')
plt.plot(T, M, lw=5, label='Analytical solution')
plt.legend()
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.show()
```

4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

susceptibility_MC = np.loadtxt('M(RT)^2/OUTPUT/susceptibility.dat') 
susceptibility_GIBBS = np.loadtxt('GIBBS/OUTPUT/susceptibility.dat')
plt.figure(figsize=(10, 6))
plt.errorbar(susceptibility_MC[:,0], susceptibility_MC[:,1] , susceptibility_MC[:,2], fmt = "ko", ms = 10, label='MC simulation')
plt.errorbar(susceptibility_GIBBS[:,0], susceptibility_GIBBS[:,1] , susceptibility_GIBBS[:,2], fmt='rx', ms=10, mew=2, label='Gibbs simulation')
plt.plot(T, X, lw=5, label='Analytical solution')
plt.legend()
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.show()
```

As showed by graphs, both algorithms successfully reproduce the thermodynamic properties of the system, yielding results that are consistent with the exact solution in the one-dimensional case. #### Acceptance Probability

As the temperature decreases, the acceptance probability for Monte Carlo moves reduces. At high temperatures, moves that increase energy are more readily accepted due to the dominance of entropy. However, at low temperatures, the system tends to favor low-energy configurations, and the acceptance rate drops significantly. To counteract this, it might be necessary to increase the number of Monte Carlo steps per temperature to ensure sufficient exploration of the configuration space. Nevertheless, as shown by the simulation plots, the results still follow the analytical solution even at low temperatures, indicating that the sampling remains effective. In contrast, for the Gibbs sampling method, the acceptance rate is always 1 by construction, as configurations are drawn directly from the conditional probability distribution.
