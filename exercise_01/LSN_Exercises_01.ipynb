In Statistics we cannot prove that some random events are drawn from a particular distribution (Note, in fact, that such hypothesis is false: pseudo-random numbers are drawn from a deterministic algorithm!); we can try to estimate the probability that $r$ **are not** drawn from a uniform distribution. If this probability is low, we can safely reject this last hypothesis.
Pearson's cumulative test statistic $\chi^2$ is a measure of the error between observations, $O_i$, and expected values, $E_i$:
$\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}$
The numerator is a squared distance between observations, $O_i$, and expected values, $E_i$, and thus should be compared (at the denominator) with the expected squared fluctuations (variance) of the relative distribution. Why variance = $E_i$? 
The probability of getting exactly $k$ (independent) successes, each one with probability $p$, in $n$ trials is given by the Binomial distribution (see <a href="https://en.wikipedia.org/wiki/Binomial_distribution">this Wikipedia link</a>):
$Pr(X=k) = {{n}\choose{k}} p^k (1-p)^{n-k} = \frac{n!}{k! (n-k)!} p^k (1-p)^{n-k}$
The average of the Binomial distribution is $\langle X \rangle = np$, the variance is: $\sigma^2 = np (1-p)$. Thus if $p$ is small we have that $\sigma^2 \simeq np$, that is $\sigma^2 \simeq \langle X \rangle$ and this explains the denominator $E_i$ in $\chi^2$

### <span style="color:blue">Computing statistical uncertainties: the blocking method</span>
- A fundamental step in every Monte Carlo calculation is the estimation of its statistical uncertainty, **it's a must!** Your task would not be completed if you had not calculated it.

In computing the statistical uncertainties, i.e. the standard deviation of the mean, you need an estimation of the variance $\sigma^2$, the second central moment. First of all note that the variance can be computed from
$\sigma^2_A := \langle (A-\langle A\rangle)^2\rangle = \langle A^2 \rangle -\langle A\rangle^2 $
What is the best way to estimate $\sigma^2_A$? Imagine that your Monte Carlo estimation of $A$ is obtained from a calculation which uses $M$ Monte Carlo "steps" (intentionally, here I am generic because what is a single "step" in a Monte Carlo calculation strictly depends on the specific calculation); you can always divide such $M$ Monte Carlo "steps" in $N$ blocks, with $N<M$. In each block, you can use your $M/N$ Monte Carlo "steps" to obtain an estimate of $A$, let me call it $A_i$ with $i=1,N$, and then you have also $A^2_i$ with $i=1,N$.
At this point everything becomes clear:
$$ \langle A^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 $$
and finally the statistical uncertainty with $N$ :
$$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N A^2_i - \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 \right]} $$


- <font color="blue">In a Monte Carlo calculation, you can use the arguments above to outputting an estimate of A and its statistical uncertainty on the fly, during your calculation. You will be able to do this every $M/N$ steps, and thus, in the end, $N$ times</font>. Note that after the first $M/N$ Monte Carlo steps, at the end of the first block, your estimation of the uncertainty is not computable, so set it to zero and compute it only from the second block. **Question**: $N$ can be chosen in many ways ... what are the limitations regarding this choice?

#### Exercise 01.2

To show the convergence of limit theorems I draw numbers for uniform, exponential and Lorentz distribution and take the average were $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, being $x_i$ the numbers drawn. I save values for $N=1, 2, 10, 100$ to appreciate the progressive convergenve of the distributions. For each value of $N$ the process is iterated over $n = 10^5$ iterations. Below there are showed hystograms of $S_N$ and I try to fit the case $N=100$ with a Gaussian for standard and exponential dices, whereas I use a Cauchy-Lorentz distribution for the last case.

```python

def gaussian(x, mu, sigma, A):
    return A * norm.pdf(x, mu, sigma)

def lorentzian_func(x, x0, gamma, A):
    return A * cauchy.pdf(x, x0, gamma)

```

```python

N_values = [1, 2, 10, 100]

for N in N_values:
    data = np.loadtxt(f'es_01.2/output_N{N}.dat')
    
    uniform = data[:, 0]
    exponential = data[:, 1]
    lorentzian = data[:, 2]

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    axes[0].hist(uniform, bins=50, density=True, alpha=0.6, color='royalblue', edgecolor='black', label="Simulazione")
    axes[0].set_title(f'Uniform - N={N}', fontsize=14)
    axes[0].set_xlabel('$S_N$', fontsize=12)
    axes[0].set_ylabel('Density', fontsize=12)
    
    axes[1].hist(exponential, bins=50, density=True, alpha=0.6, color='crimson', edgecolor='black', label="Simulazione")
    axes[1].set_title(f'Exponential - N={N}', fontsize=14)
    axes[1].set_xlabel('$S_N$', fontsize=12)

    axes[2].hist(lorentzian, bins=50, density=True, alpha=0.5, color='darkgreen', edgecolor='black', label="Simulazione", range=(-20, 20))
    axes[2].set_title(f'Lorentz - N={N}', fontsize=14)
    axes[2].set_xlabel('$S_N$', fontsize=12)
    
    #  N = 100
    if N == 100:
        # UNIFORM
        x_range_unif = np.linspace(min(uniform), max(uniform), 1000)
        hist_unif, bins_unif = np.histogram(uniform, bins=50, density=True)
        bin_centers_unif = 0.5 * (bins_unif[1:] + bins_unif[:-1])
        
        popt_unif, _ = curve_fit(gaussian, bin_centers_unif, hist_unif, p0=[np.mean(uniform), np.std(uniform), 1])
        mu_u, sigma_u, A_u = popt_unif
        axes[0].plot(x_range_unif, gaussian(x_range_unif, *popt_unif), 'b-', lw=3)
        display(Math(f"[Uniform Fit] \\mu\\ = {mu_u:.4f}, \\sigma\\ = {sigma_u:.4f}"))

        # EXPONENTIAL
        x_range_exp = np.linspace(min(exponential), max(exponential), 1000)
        hist_exp, bins_exp = np.histogram(exponential, bins=50, density=True)
        bin_centers_exp = 0.5 * (bins_exp[1:] + bins_exp[:-1])

        popt_exp, _ = curve_fit(gaussian, bin_centers_exp, hist_exp, p0=[np.mean(exponential), np.std(exponential), 1])
        mu_e, sigma_e, A_e = popt_exp
        axes[1].plot(x_range_exp, gaussian(x_range_exp, *popt_exp), 'r-', lw=3)
        display(Math(f"[Exponential Fit] \\mu\\ = {mu_e:.4f}, \\sigma\\ = {sigma_e:.4f}"))

        # LORENTZIAN
        filtered_lor = lorentzian[np.abs(lorentzian) < 20]
        x_range_lorentz = np.linspace(-20, 20, 1000)
        hist_lor, bins_lor = np.histogram(filtered_lor, bins=50, density=True)
        bin_centers_lor = 0.5 * (bins_lor[1:] + bins_lor[:-1])

        p0_lorentz = [np.median(filtered_lor), 
                    (np.percentile(filtered_lor, 75) - np.percentile(filtered_lor, 25)) / 2,
                    1]

        popt_lorentz, _ = curve_fit(lorentzian_func, bin_centers_lor, hist_lor, p0=p0_lorentz )
        x0, gamma, A_l = popt_lorentz
        axes[2].plot(x_range_lorentz, lorentzian_func(x_range_lorentz, *popt_lorentz), 'g-', lw=3)
        display(Math(f"[Lorentzian Fit] \\mu\\ = {x0:.4f}, \\gamma\\ = {gamma:.4f}"))


    plt.tight_layout()
    plt.show()
```

The Gaussian fits for the uniform and exponential samples confirm the central limit theorem, as both have finite variance and their sample means converge to a normal distribution. In contrast, the Cauchy distribution, having infinite variance, does not follow the central limit theorem, but instead remains stable under summation, in line with the generalized central limit theorem for stable distributions.

#### Exercise 01.3


In this simulation, I estimate the value of $\pi$ using the **Buffon's Needle** experiment.

I use the following parameters:

- `L = 1.0` â€ƒ Length of the needle  
- `d = 1.5` â€ƒ Distance between the parallel lines  
- `M = 100000` â€ƒ Total number of throws  
- `N = 100` â€ƒ Number of blocks  

For each throw, I randomly extract:

- $x$: the distance of the needle's center from the nearest line, uniformly distributed in $[0, d/2]$.
- $\theta$: the angle between the needle and the parallel lines.  
  $\theta$ is generated without using $\pi$ by uniformly sampling a point $(x', y')$ in the upper half of the unit circle $(y' \geq 0)$, and computing:
  
  $$
  \theta = \arccos\left(\frac{x'}{\sqrt{x'^2 + y'^2}}\right)
  $$

The needle intersects a line if:

$$
x \leq \frac{L}{2} \cdot \sin(\theta)
$$

The estimation of $\pi$ is then given by:

$$
\pi \approx \frac{2L}{d} \cdot \frac{M}{\text{Number of hits}}
$$

The plot below shows the progressive average of $(\langle \pi_{\text{estimation}} \rangle - \pi)$ with statistical errors as error bars.


```python

pi = np.loadtxt('es_01.3/pi_estimate.dat')

plt.figure(figsize=(12,6))
plt.errorbar(pi[:,0]*1e-3, pi[:,1] - math.pi,yerr= pi[:,2],  capsize=3, lw=2, label='Blocking average')
plt.xlabel('#blocks')
plt.ylabel(r'$<\pi_{estimation}>-\pi$')
plt.grid(True)
plt.hlines(0, 0, 100, colors='r', linestyles='dashed', label='Expected value',lw=3)
plt.legend()
plt.show()
```

The estimation of $\pi$ from the graph can be seen quickly converging to the real value.
